# Solution

```python
def lonelyinteger(a):
    if not a:
        return None  # Return None for an empty input list

    a.sort()  # Sort the input list in-place
    i = 0

    while i < len(a) - 1:
        if a[i] == a[i + 1]:
            i += 2  # Skip the pair of equal elements
        else:
            return a[i]  # Found the lonely integer

    # If no lonely integer is found, return the last element
    return a[-1]

# Example usage:
arr = [1, 2, 3, 4, 1, 2, 3]
result = lonelyinteger(arr)
print("Lonely integer:", result)
```
