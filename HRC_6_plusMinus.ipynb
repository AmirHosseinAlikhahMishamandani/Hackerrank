# Solution

```python
import math
import os
import random
import re
import sys

def plusMinus(arr):
    Positive_proportion = 0
    Negative_proportion = 0
    zeros_proportion = 0

    for num in arr:
        if num > 0:
            Positive_proportion += 1
        elif num < 0:
            Negative_proportion += 1
        elif num == 0:
            zeros_proportion += 1

    print("{:.6f}\n{:.6f}\n{:.6f}".format(Positive_proportion/len(arr), Negative_proportion/len(arr), zeros_proportion/len(arr)))

# 0 < n <= 100
n = random.randint(1,100)
# -100 <= arr[i] <= 100
arr = [random.randint(-100,100) for _ in range(n)]

print(f'The given array is:\n\n{arr}\n')

plusMinus(arr)
```
