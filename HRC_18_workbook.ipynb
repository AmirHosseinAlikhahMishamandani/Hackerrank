# Solution

```python
import math
import os
import random
import re
import sys

def workbook(n, k, arr):
    specials = 0
    page = 1
    for problems_in_chapter in arr:
        for problem_in_chapter in range(1, problems_in_chapter + 1):
            if problem_in_chapter == page:
                specials += 1
            if problem_in_chapter % k == 0 or problem_in_chapter == problems_in_chapter:
                page += 1

    return specials

n = 5
k = 3
arr = [4, 2, 6, 1, 10]

print(f'The given array with {n} chapters and each page with maximum {k} problems is having {workbook(n, k, arr)} special probelms.')
```
