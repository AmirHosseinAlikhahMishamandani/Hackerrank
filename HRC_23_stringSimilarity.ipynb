# Solution

```python
# z_array

def stringSimilarity(s):
    n = len(s)
    z = [0] * n
    l, r, sum_similarity = 0, 0, 0

    for i in range(1, n):
        if i <= r:
            z[i] = min(r - i + 1, z[i - l])

        while i + z[i] < n and s[z[i]] == s[i + z[i]]:
            z[i] += 1

        if i + z[i] - 1 > r:
            l, r = i, i + z[i] - 1

        sum_similarity += z[i]

    return sum_similarity + n

s = 'ababaa'

print(f'{stringSimilarity(s)}')
```
