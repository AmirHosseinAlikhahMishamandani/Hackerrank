# Solution

```python
import math
import os
import random
import re
import sys

def aVeryBigSum(ar):
    return sum(ar)

# 1 <= n <= 10
# 0 <= ar[i] <= 10^10
n = random.randint(1,10)
ar = [random.randint(0, 10**10) for _ in range(n)]

print(f'The given array is: {ar}\n')
print(f'The sum of the numbers in the array is: {aVeryBigSum(ar)}\n')
```
