# Solution

```python
import math
import os
import random
import re
import sys

def birthdayCakeCandles(candles):
  return candles.count(max(candles))

# 1 <= n <= 10**5
n = random.randint(1,10**5)
# 1 <= candles[i] <= 10**7
candles = [random.randint(1,10**7) for _ in range(n)]

print(f'The list of candles are as follows:\n\n{candles}\n')

print(f'The number of the tallest candle with the height of {max(candles)} is:\n{birthdayCakeCandles(candles)}')
```
