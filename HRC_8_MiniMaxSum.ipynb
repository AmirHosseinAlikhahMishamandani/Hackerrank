# Solution

```python
import math
import os
import random
import re
import sys
from itertools import combinations

def miniMaxSum(arr):
    combinations_4 = list(combinations(arr, 4))
    sum_list = []
    for comb in combinations_4:
        sum_list.append(sum(comb))
    print(f'{min(sum_list)} {max(sum_list)}')

# Max Length of an array is 5
n = 5
# 1 <= arr[i] <= 10**9
arr = (random.randint(1,10**9) for _ in range(5))

print('The min & max are:\n')
miniMaxSum(arr)
```
