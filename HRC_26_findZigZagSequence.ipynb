# Solution

```python
def findZigZagSequence(a, n):
    a.sort()
    mid = int((n - 1)/2) # <------- Change Here
    a[mid], a[n-1] = a[n-1], a[mid]

    st = mid + 1
    ed = n - 2 # <------- Change Here
    while(st <= ed):
        a[st], a[ed] = a[ed], a[st]
        st = st + 1
        ed = ed - 1 # <------- Change Here

    for i in range (n):
        if i == n-1:
            print(a[i])
        else:
            print(a[i], end = ' ')
    return

n = 7
a = [1,2,3,4,5,6,7]

findZigZagSequence(a, n)
```
