```python
import math
import os
import random
import re
import sys

def boardCutting(cost_y, cost_x):
    cost_y.sort(reverse=True)
    cost_x.sort(reverse=True)
    y_index, x_index = 0, 0
    H_seg, V_seg = 1, 1
    minCost = 0
    MOD = 10**9 + 7

    while y_index < len(cost_y) and x_index < len(cost_x):
        if cost_y[y_index] >= cost_x[x_index]:
            minCost += cost_y[y_index] * V_seg
            y_index += 1
            H_seg += 1
        else:
            minCost += cost_x[x_index] * H_seg
            x_index += 1
            V_seg += 1
        minCost %= MOD

    while y_index < len(cost_y):
        minCost += cost_y[y_index] * V_seg
        y_index += 1
        minCost %= MOD

    while x_index < len(cost_x):
        minCost += cost_x[x_index] * H_seg
        x_index += 1
        minCost %= MOD

    return minCost


# 1 <= q <= 20
q = random.randint(1,20)
# 2 <= m,n <= 1000000
m = random.randint(2,1000000)
n = random.randint(2,1000000)
# 0 <= cost_y[i], cost_x[j] <= 10**9
cost_y = [random.randint(0,10**9) for _ in range(m)]
cost_x = [random.randint(0,10**9) for _ in range(n)]

print(f'The min Cost of the given list is: {boardCutting(cost_y, cost_x)}')
```
