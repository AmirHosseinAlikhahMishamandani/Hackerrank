# Solution

```python
import math
import os
import random
import re
import sys

def compareTriplets(a, b):
    scores = [0, 0]  # Alice, Bob
    for i in range(3):
        if a[i] > b[i]:
            scores[0] += 1
        elif a[i] < b[i]:
            scores[1] += 1
    return scores

# 1 <= a[i], b[i] <= 100
# [problem clarity, originality, and difficulty]
a = [random.randint(1, 100) for _ in range(3)]
b = [random.randint(1, 100) for _ in range(3)]

print(f'Alice: {a}\n')
print(f'Bob: {b}\n')

print(f'The Alice and Bobs comparison scores are: {compareTriplets(a, b)}, respectively')
```
