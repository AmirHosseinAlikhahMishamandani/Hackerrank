```python
import math
import os
import random
import re
import sys

def countApplesAndOranges(s, t, a, b, apples, oranges):
    sam_apple = 0
    sam_orange = 0
    for d_a in apples:
        if a + d_a >= s and a + d_a <= t:
            sam_apple += 1
    for d_o in oranges:
        if b - d_o >= s and b - d_o <= t:
            sam_orange += 1
    print(f'{sam_apple}\n{sam_orange}')

# 1 <= s,t,a,b,m,n <= 10**5
m = random.randint(1, 10**5)
n = random.randint(1, 10**5)
a = random.randint(1, 10**5)
b = random.randint(a, 10**5)
# a < s < t < b
s = random.randint(a, b)
t = random.randint(s, b)
# -10^5 <= d <= 10^5
apples = [random.randint(-10**5,10**5) for _ in range(m)]
oranges = [random.randint(-10**5,10**5) for _ in range(n)]

print(f'The apples and oranges that fall on Sam house are:\n')
countApplesAndOranges(s, t, a, b, apples, oranges)
```
