# Solution

```python
def caesarCipher(s, k):
    result = ""

    for char in s:
        if char.isalpha():
            # Shift character based on its case
            start = ord('a') if char.islower() else ord('A')
            result += chr((ord(char) - start + k) % 26 + start)
        else:
            # Non-alphabetical characters are added as is
            result += char

    return result

s = 'Always-Look-on-the-Bright-Side-of-Life'
k = 5

caesarCipher(s, k)
```
