# Solution

```python
import math
import os
import random
import re
import sys

def twoArrays(k, A, B):
    A.sort()
    B.sort(reverse=True)
    C = [A[i] + B[i] for i in range(len(A))]
    if min(C) >= k:
        return 'YES'
    else:
        return 'NO'

# 1 <= q <= 10
q = random.randint(1,10)
# 1 <= n <= 1000
n = random.randint(1,1000)
# 1 <= 10**9
k = random.randint(1,10**9)
# 0 <= A[i], B[i] <= 10**9
A = [random.randint(0,10**9) for _ in range(n)]
B = [random.randint(0,10**9) for _ in range(n)]

print(f'The result of the twoArrays({k}, A, B): is {twoArrays(k, A, B)}')
```
