# Solution

```python
import math
import os
import random
import re
import sys
import numpy as np

def diagonalDifference(arr):
    L2R_diag_sum = 0
    R2L_diag_sum = 0
    MatSize = len(arr)

    for i in range(MatSize):
        L2R_diag_sum += arr[i][i]
        R2L_diag_sum += arr[i][MatSize - 1 - i]

    result = abs(L2R_diag_sum - R2L_diag_sum)

    return result

# n is not given by the problem!
n = random.randint(1,10)
# -100 <= arr [i][j] <= 100
arr = np.random.randint(-100, 100 + 1, size=(n, n))

print(f'The given matrix is:\n {arr}\n')

print(f'The abs(primary diag - secondary diag = {diagonalDifference(arr)})\n')
```
