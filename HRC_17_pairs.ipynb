```python
from itertools import combinations
def pairs(k, arr):
    init_list = list(combinations(arr, 2))
    result = [abs(init_list[i + 1][0] - init_list[i][1]) for i in range(len(init_list) - 1)]
    return result.count(k)

k = 2
arr = [1, 5, 3, 4, 2]
pairs(k, arr)
```
