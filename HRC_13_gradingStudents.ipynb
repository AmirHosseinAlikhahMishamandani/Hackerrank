```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'gradingStudents' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts INTEGER_ARRAY grades as parameter.
#

def gradingStudents(grades):
    final_grades = []
    for grd in grades:
        if (grd + 1) % 5 == 0 and grd >= 38:
            final_grades.append(grd + 1)
        elif (grd + 2) % 5 == 0 and grd >= 38:
            final_grades.append(grd + 2)
        else:
            final_grades.append(grd)

    return final_grades


# 1 <= n <= 60
n = random.randint(1,60)
# 0 <= grades[i] <= 100
grades = [random.randint(0,100) for _ in range(n)]

print(f'The given list of grades:\n\n{grades}')

print(f'\nUpdated list of gardes:\n\n{gradingStudents(grades)}')
```
