# Solution

```python
def migratoryBirds(arr):
  Brd = sorted(set(arr), reverse=True)
  Max_Id = 0
  Max_Cnt = 0
  for Id in Brd:
    if arr.count(Id) >= Max_Cnt:
      Max_Cnt = arr.count(Id)
      Max_Id = Id

  return Max_Id

arr = [1, 2, 3, 4, 5, 4, 3, 2, 1, 3, 4]

migratoryBirds(arr)
```
