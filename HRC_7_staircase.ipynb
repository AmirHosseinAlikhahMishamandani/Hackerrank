# Solution

```python
import math
import os
import random
import re
import sys

def staircase(n):
    for i in range(1, n+1):
        out = ' ' * (n-i) + '#' * i
        print(out)

# 0 < n <= 100
n = random.randint(1,100)

print(f'The staircase for n = {n}\n')
staircase(n)
```
