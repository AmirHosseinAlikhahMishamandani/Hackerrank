```python
def breakingRecords(scores):
  max_record = scores[0]
  min_record = scores[0]
  max_record_change = 0
  min_record_change = 0
  for rec in scores:
    if rec > max_record:
      max_record = rec
      max_record_change += 1
    elif rec < min_record:
      min_record = rec
      min_record_change += 1
  return [max_record_change, min_record_change]

scores = [10, 5, 20, 20, 4, 5, 2, 25, 1]

breakingRecords(scores)
```
