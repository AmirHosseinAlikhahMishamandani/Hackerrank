# Solution

```python
import math
import os
import random
import re
import sys

def simpleArraySum(ar):
    return sum([num for num in ar])

# 0 < n, ar[i] <= 1000
n = random.randint(1,1000)
random_list = [random.randint(1, 1000) for _ in range(n)]

print(f'List of the random integers: {random_list}\n')

print(f'The sum of the given list is:\n\n {simpleArraySum(random_list)}')
```
